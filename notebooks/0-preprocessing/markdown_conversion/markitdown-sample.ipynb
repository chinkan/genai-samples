# Image to Text Example

For image, we need to use llm model to generate text from image.

```python
%pip install openai
```

```python
# Uncomment to Use OpenAI
# from openai import OpenAI
# client = OpenAI()

# Azure OpenAI
from openai import AzureOpenAI
client = AzureOpenAI()

model = "gpt-4o-mini"

from markitdown import MarkItDown
md = MarkItDown(llm_client=client, llm_model=model)
# result = md.convert("../../../data/raw/sample.webp") #webp is not supported
result = md.convert("../../../data/raw/sample.jpg")

from IPython.display import Image
display(Image(filename='../../../data/raw/sample.jpg'))
print(result.text_content)
```
